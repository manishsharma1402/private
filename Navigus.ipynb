```python
import tkinter
from tkinter import *
from tkinter import messagebox
```

```python
root = Tk()
root.title("Navigus")

def emergency_click():
    myLabel=Label(root, fg="red",text=("Students Area"))
    myLabel.grid(columnspan=3, padx=10, pady=10 , )
    
                   
    myLabel1 = Label(root, text="Course Name:")
    myLabel1.grid() 
                   
    enter_temp = Entry(root, width= 35 , borderwidth=5)
    enter_temp.grid(row=2,column=1, columnspan=3, padx=10, pady=10)
    
    def pop_button():
        response=messagebox.askquestion("Student","Is this final Course Name")
        
        if response=="yes":
            Label(root,text="Your course name is saved.",fg="red").grid()
            messagebox.showinfo("Student","your course is saved successfully")
        else:
            Label(root,text="Enter the valid course name",fg="green").grid()
        
    
    pop_button=Button(root, text="Save Course", command=pop_button,fg="Yellow",bg="Black")
    pop_button.grid(column=1)
    
    def submit_button():
        messagebox.showinfo("Student","You have successfully enrolled in this course.Now you can attempt quiz." )
        root.destroy()
    
    pop_button=Button(root, text="Save Info", command=submit_button,fg="white",bg="Black",padx=10, pady=10)
    pop_button.grid(column=0)
    
    

    
   

    
    
def regular_click():
    myLabel=Label(root,fg="red", text=("Teachers Working Area"))
    myLabel.grid(columnspan=3, padx=10, pady=10)
    
    myLabel2 = Label(root, text="Enter the Course Name:")
    myLabel2.grid() 
    
    
                   
    enter_place = Entry(root, width= 35 , borderwidth=5)
    enter_place.grid(row=2,column=1, columnspan=3, padx=10, pady=10)
    
    def pop1_button():
        response=messagebox.askquestion("Teachers","Are you ready to create quiz:")
        
        if response=="yes":
            Label(root,text="Now you can create a quiz!",fg="red").grid()
            messagebox.showinfo("Regular Services","You can add , remove , edit or delete any course any time.")
            
        else:
            Label(root,text="Come back when you want to create a quiz.",fg="green").grid()
        
    
    pop1_button=Button(root, text="Save", command=pop1_button,fg="Yellow",bg="Black")
    pop1_button.grid(row=3,column=1)
    
    myLabel3 = Label(root, text="Do you want to create quiz for multiple subjects?")
    myLabel3.grid() 
    
    
                     
    enter_place = Entry(root, width= 35 , borderwidth=5)
    enter_place.grid(row=4,column=1, columnspan=3, padx=10, pady=10)
    
    def pop2_button():
        response=messagebox.askquestion("Teachers","Do you want to create quiz for multiple subjects?")
        
        if response=="yes":
            Label(root,text="Okay! you have to go one by one .",fg="red").grid()
            messagebox.showinfo("Teachers","You will be redirected to quiz 1")
            
        else:
            Label(root,text="You can go for only One quiz.",fg="green").grid()
            
    pop2_button=Button(root, text="Save Data", command=pop2_button,fg="Yellow",bg="Black")
    pop2_button.grid(row=5,column=1)
    
    
   
    
    def submit_button():
        messagebox.showinfo("Teachers","You will be redirected to Quiz creation." )
        root.destroy()
            
       
    pop_button=Button(root, text="Save Info", command=submit_button,fg="white",bg="Black",padx=10, pady=10)
    pop_button.grid(row=8,column=0)
   


emergencyButton=Button(root, text ="Students" ,font=18, padx=50,pady=20 ,bd=4 ,fg="black",bg="white" ,command=emergency_click)
emergencyButton.grid(row=0 ,column=0)

regularButton=Button(root, text ="Teachers",font=18,padx=50,pady=20 , bd=4 ,fg="black",bg="white" ,command=regular_click)
regularButton.grid(row=0 , column=1)




root.mainloop()

    
    
    
```
